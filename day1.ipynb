```
x=[2,9,7,3]
y=[7,5,3,6]

plt.xlabel("Distance")  #x axis
plt.ylabel("Time")  #y axis
plt.plot(x,y,color='black',label='railway track')   #it will draw a graph
plt.scatter(x,y,s=200,color="red",label='railway station',marker='*') #it will plot dots 
plt.bar(x,y,label='height of railway station') #to plot bar charts
plt.grid(color='green') #will draw a grid
plt.legend() #it will show label in plot
plt.show()  #will draw graph
```
